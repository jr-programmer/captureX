The document titled "Value Creation in Sustainable Energy Transition Using Reinforcement Learning" discusses the application of Reinforcement Learning (RL) to optimize decision-making in the context of energy companies navigating the transition from fossil fuels to renewable energy sources. Below is a summary of the key points:

Abstract:
The paper addresses the global challenge of climate change and how energy companies need to balance profit, CO2 emission reduction, and competitive advantage during the transition to renewable energy. It introduces a multi-objective Sequential Decision Making (SDM) framework that models the dynamic energy landscape up to 2050. The framework explores decision strategies for investment in hydrocarbons, renewables, and CO2 reduction, using Reinforcement Learning (RL) to optimize these decisions under uncertainty.

Introduction:
The energy sector is a major contributor to global CO2 emissions, and transitioning to renewable energy is essential but difficult. Energy companies face a dilemma between acting too late and losing competitive value or acting too early and facing technological risks. This research emphasizes viewing the energy transition as a Sequential Decision-Making process and proposes RL as a tool to improve decision-making in the energy sector.

Design of the Sequential Decision-Making Framework:
The framework focuses on modeling an energy company's decisions over a 25-year period, balancing investments between hydrocarbons, renewables, and CO2 capture and storage (CCS) projects. The model considers uncertainties such as energy prices, production costs, and technological advancements. The company aims to maximize Net Present Value (NPV), minimize CO2 emissions, and enhance Competitive Advantage (CA).

Methods:
The research introduces RL and explains how it is well-suited for sequential decision-making problems. The Q-Learning Algorithm and Deep Q Networks (DQN) are used to train a neural network model that optimizes investment strategies. RL is highlighted as an effective tool to manage complex decisions with large state spaces, where traditional methods like dynamic programming may fail.

Results and Discussion:
The model architecture involves training a deep neural network to calculate Q-values (cumulative rewards) for various state-action pairs. The training results show that RL can balance multiple conflicting objectives, such as maximizing NPV while reducing CO2 emissions. Comparisons with other benchmark strategies demonstrate the superiority of RL in achieving better overall performance across all objectives.

The results also suggest a shift in investment strategy from hydrocarbons to renewables over time, indicating that while hydrocarbons are essential in the short term, renewables become increasingly important for long-term sustainability.

Conclusions:
Reinforcement Learning effectively enhances decision-making in energy companies by dynamically adjusting to market conditions and environmental objectives.
The energy transition will require continued investments in hydrocarbons for the next decade, but renewable investments will become dominant in the long term.
Flexibility in decision-making is key to turning uncertainties into opportunities during the energy transition.
Limitations & Future Work:
The study acknowledges several limitations, including the high number of uncertainties in the model, the simplicity of the scaling method used, and the unrealistic budget model. Future work is suggested to address these issues, such as incorporating broader decision-making actions and refining the model to better reflect real-world investment strategies.

This research provides a valuable framework for energy companies to strategically manage the transition to sustainable energy using RL techniques, while balancing financial performance, environmental impact, and competitive positioning.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Set random seed for reproducibility
np.random.seed(42)
```

```python
# Years
years = np.arange(2025, 2050)
```

```python
# Dummy Data: Investments in million USD
hydrocarbon_investment = np.random.uniform(500, 1500, len(years))
renewable_investment = np.random.uniform(500, 1500, len(years))
co2_capture_investment = np.random.uniform(200, 800, len(years))
```

```python
# Dummy Data: Simulated objectives (scaled for simplicity)
npv = np.cumsum(hydrocarbon_investment * 0.05 + renewable_investment * 0.07 - co2_capture_investment * 0.03)
co2_emissions = np.cumsum(hydrocarbon_investment * 0.2 - co2_capture_investment * 0.5)
competitive_advantage = np.cumsum(renewable_investment * 0.08 - hydrocarbon_investment * 0.04)
```

```python
# Create DataFrame to store dummy data
df = pd.DataFrame({
    'Year': years,
    'Hydrocarbon Investment (M USD)': hydrocarbon_investment,
    'Renewable Investment (M USD)': renewable_investment,
    'CO2 Capture Investment (M USD)': co2_capture_investment,
    'NPV': npv,
    'CO2 Emissions': co2_emissions,
    'Competitive Advantage': competitive_advantage
})
```

```python
# Display the data
print(df)
```

```python
# Plot the investment strategy over time
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Hydrocarbon Investment (M USD)'], label='Hydrocarbon Investment', color='red', marker='o')
plt.plot(df['Year'], df['Renewable Investment (M USD)'], label='Renewable Investment', color='green', marker='o')
plt.plot(df['Year'], df['CO2 Capture Investment (M USD)'], label='CO2 Capture Investment', color='blue', marker='o')
plt.xlabel('Year')
plt.ylabel('Investment (Million USD)')
plt.title('Investment Strategy Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot the NPV, CO2 Emissions, and Competitive Advantage over time
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['NPV'], label='NPV', color='purple', marker='o')
plt.plot(df['Year'], df['CO2 Emissions'], label='CO2 Emissions', color='orange', marker='o')
plt.plot(df['Year'], df['Competitive Advantage'], label='Competitive Advantage', color='cyan', marker='o')
plt.xlabel('Year')
plt.ylabel('Objective Value')
plt.title('NPV, CO2 Emissions, and Competitive Advantage Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
